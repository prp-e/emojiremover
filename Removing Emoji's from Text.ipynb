# Removing Emoji's from a text

## Needed Library

```python
from hazm import word_tokenize 
```

As our text is mostly in Persian, I used this library made specificly for Persian language called "hazm". You also can use *NLTK*. 

## Creating a useful list of emoji's

First, we just open the `emojis.txt` file. Then, we extract all emoji's (in this case, all emojis from Telegram messenger) as a large string. 

```python
emoji_list = open("emojis.txt")
emoji_list = emoji_list.readlines()
emoji_list = emoji_list[0]

print(emoji_list)
```

Now, we create an array of emojis like this:

```python
emoji_list_functional = [] 

for emoji_char in emoji_list:
    emoji_list_functional.append(emoji_char)
    
print(emoji_list_functional)
```

Now, we need a sample text, including some emojis. Then, we tokenize the text. 

```python
sample_text = "وای 🤦🏻‍♀️ فکر کنم گند زدم! ولی عیبی نداره خودمونو عشقه 😍🥰"
sample_text = word_tokenize(sample_text)

print(sample_text)
```

And after that, we need to clean our text. This part of our code (which can be found in `emojiremover.py` script as well), does the job. 


```python
clean_text = []

for word_token in sample_text:
    for emoji_char in emoji_list_functional:
        word_token = word_token.replace(emoji_char, '')
    clean_text.append(word_token)
```

This is a nested `for` loop. First, it explores the array, then compares each member of the `sample_text` with members of the `emojis_list_functional` list. Then, it replaces each emoji with `''` characters. 

```python
print(f'Input: {sample_text}') 
print(f'Output: {clean_text}')
```

Now, you can see how it acted above. Now, I made it better, by combining everything in a function like the function below:

```python
def remove_emojis(text, emojis="emojis.txt"):
    ''' Removes emojis from a given text! '''
    emoji_list = open(emojis)
    emoji_list = emoji_list.readlines()
    emoji_list_functional = []
    for emoji_char in emoji_list[0]:
        emoji_list_functional.append(emoji_char)
    text = word_tokenize(text)
    clean_text = []
    for word_token in text:
        for emoji_char in emoji_list_functional:
            word_token = word_token.replace(emoji_char, '')
        clean_text.append(word_token)
    return clean_text
```
